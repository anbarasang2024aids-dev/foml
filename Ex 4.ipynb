```python
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
Y = np.array([0, 0, 0, 1])
weights = np.array([0.3, -0.2])
bias = -0.4
learning_rate = 0.2
```

```python
import numpy as np
```

```python
def activation(net_input):
  return 1 if net_input >= 0 else 0
```

```python
epoch = 0
while True:
    print(f"\nEpoch {epoch + 1}")
    error_count = 0
    for i in range(len(X)):
        net_input = np.dot(X[i], weights) + bias
        output = activation(net_input)
        error = Y[i] - output

        print(f"Input: {X[i]}, Output: {output}, Desired: {Y[i]}, Error: {error}")

        if error != 0:
            weights += learning_rate * error * X[i]
            bias += learning_rate * error
            print(f"Updated weights: {weights}, Updated bias: {bias}")
            error_count += 1
        else:
            print("No change.")

    epoch += 1
    if error_count == 0:
        print("\nTraining complete.")
        break
```

```python
print(f"\nFinal weights: {weights}")
print(f"Final bias: {bias}")
```
